# Download album covers

```python
from time import sleep
def download_save_image(row):
    urllib.request.urlretrieve(row['album_cover_url'], "./images/"+str(row['id'])+".jpg")
    sleep(5)    
```

```python
df.apply(lambda x:download_save_image(x),axis=1)
```
